### Exercise 5: Subplot Layout

```python
import matplotlib.pyplot as plt
import numpy as np


x = np.arange(1, 6)
y = x ** 2
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 9]
scatter_x = [1, 2, 3, 4, 5]
scatter_y = [2, 4, 6, 8, 10]
labels = ["Python", "Java", "C++", "JavaScript"]
sizes = [35, 25, 20, 20]


fig, axs = plt.subplots(2, 2, figsize=(10, 8))


axs[0, 0].plot(x, y, marker='o', linestyle='-', color='b')
axs[0, 0].set_title("Line Plot")
axs[0, 0].set_xlabel("X-axis")
axs[0, 0].set_ylabel("Y-axis")


axs[0, 1].bar(categories, values, color=['r', 'g', 'b', 'c'])
axs[0, 1].set_title("Bar Chart")


axs[1, 0].scatter(scatter_x, scatter_y, color='purple', marker='o')
axs[1, 0].set_title("Scatter Plot")
axs[1, 0].set_xlabel("X-axis")
axs[1, 0].set_ylabel("Y-axis")


axs[1, 1].pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
axs[1, 1].set_title("Pie Chart")


plt.tight_layout()


plt.show()
```
