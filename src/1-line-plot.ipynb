### Exercise 1: Line Plot Customization

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(0, 19)
y = x**2

plt.plot(x, y, marker='o', linestyle='-', color='b', label="y = x^2")

plt.title("Line Plot: y = xÂ²")

plt.xlabel("X Values")
plt.ylabel("Y Values (Squared)")

plt.grid(True)
plt.legend()

plt.show()
```
